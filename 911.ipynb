```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('911.csv')
```

```python
df
```

```python
df_plot=df[['twp','Reason','hour','month','day of week','Year']]
```

```python
df_plot['ntitle']=df['title'].str.split(':').str[0]
```

```python
df_plot
```

```python
df_plot['ntitle'].value_counts()
```

## Drop New Column again as the same as Reason

```python
df_plot.drop('ntitle',axis=1,inplace=True)
```

```python
sns.set(rc={'figure.figsize':[10,4]},font_scale=1.3)
for col in df_plot.select_dtypes(exclude=['O','bool']):
    sns.displot(df_plot[col],bins=20,kde=True,color='r')
```

```python
for col in df_plot.select_dtypes(include='O'):
        plt.title(col)
        sns.countplot(data=df_plot,x=col,palette='Accent')
        plt.show()
```

```python
df_plot
```

```python
for col in df_plot.select_dtypes(include='O'):
        plt.title(col+' & Reason')
        sns.countplot(data=df_plot,x=col,hue='Reason',palette='mako')
        plt.show()
```
