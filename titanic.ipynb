```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_plot=df[['survived','sex','age','fare','pclass','who','embark_town','alive','alone']]
```

```python
df_plot
```

### Numerical Univarite Features

```python
sns.set(rc={'figure.figsize':[10,4]},font_scale=1.3)
for col in df_plot.select_dtypes(exclude=['O','bool']):
    sns.displot(df_plot[col],bins=20,kde=True,color='r')
```

### Categrical Univarite Features

```python
for col in df_plot.select_dtypes(include='O'):
    if col=='alive':
        continue
    else:
        plt.title(col)
        sns.countplot(data=df_plot,x=col,palette='Accent')
        plt.show()
```

```python
for col in df_plot.select_dtypes(include='O'):
    if col=='alive':
        continue
    else:
        plt.title(col+' & alive')
        sns.countplot(data=df_plot,x=col,hue='alive',palette='mako')
        plt.show()
```
